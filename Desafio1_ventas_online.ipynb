#**Desafío Ventas Online**

```python
import pandas as pd
```

##**1. Lectura de datos:**

```python
ruta_datos = '/content/drive/MyDrive/Colab Notebooks/Ruta: Aprendiendo a hacer ETL (7)/Curso 4-7 Data Science: Pandas: manupulacion y transformacion de datos/Desafio 1: Ventas Online/dados_vendas_clientes.json'
datos = pd.read_json(ruta_datos)
datos.head()
```

```python
datos.info()
```

```python
datos = pd.json_normalize(datos['dados_vendas'])
```

```python
datos.head()
```

##**2. Eliminar datos en listas dentro del DataFrame:**

```python
columnas = list(datos.columns)
columnas
```

```python
datos = datos.explode(columnas[1:])
```

```python
datos.reset_index(inplace=True, drop=True)
```

```python
datos.head()
```

##**3. Verificar tipos de datos:**

```python
datos.info()
```

##**4. Identificar columnas numéricas:**

```python
datos

#La columna 'valor da compra' debe de ser numérica
```

```python
datos['Valor da compra'] = datos['Valor da compra'].apply(lambda x: x.replace('R$','').replace(',','').strip())
```

```python
import numpy as np
```

```python
datos['Valor da compra'] = datos['Valor da compra'].astype(np.float64)
```

##**5. Manipulación de un texto:**

**1. Pasar todo a minúscula:**

```python
datos['Cliente'] = datos['Cliente'].str.lower()
datos['Cliente'].head()
```

**2. Eliminar caracteres especiales:**

```python
datos['Cliente'] = datos['Cliente'].str.replace('[^a-z]',' ', regex=True)
datos['Cliente'].head()
```

##**6. Datos de tiempo:**

```python
# Transformar para el tipo datetime definiendo el formato de fecha como DD/MM/AAAA ('%d/%m/%Y')
datos['Data de venda'] = pd.to_datetime(datos['Data de venda'], format='%d/%m/%Y')
```

```python
datos
```

```python
datos.info()
```

```python
# Calcular el total recaudado en compras por cada cliente
total_compras = datos.groupby(['Cliente'])['Valor da compra'].sum()

total_compras
```
