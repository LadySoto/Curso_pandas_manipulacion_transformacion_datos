El método **apply()** solo funciona para **Series** y el meétodo **applymap()** para **DataFrame**, una **Serie** es cuando tengo una sola columna o estoy trabajando en una sola columna, pero un **DataFrame** es cuando tengo más de dos columnas

```python
datos['precio'] = datos['precio'].apply(lambda x: x.replace('$','').replace(',','').strip())
```

para acceder a varias columnas podría hacerlo de dos formas:

1. creando una lista y después trabajo con ellas

* colums_moneda = ['cuota_deposito','cuota_limpieza']
* datos[colums_moneda]

2. con el doble par de **[ ]**

* datos[ ['cuota_deposito','cuota_limpieza'] ]

```python
colums_moneda = ['cuota_deposito','cuota_limpieza']
datos[colums_moneda] = datos[colums_moneda].applymap(lambda x: x.replace('$','').replace(',','').strip())
```

Me suma la cantidad de inmuebles que estaban disponibles por cada año/mes

```python
disponibilidad_por_ano_mes = datos2.groupby(fecha_ano_mes)['lugar_disponible'].sum()
disponibilidad_por_ano_mes.head()
```

**Explicación al código:**

**cantidad_dias_por_inmueble = datos_ejemplo.groupby(' id ')[' fecha '].nunique( )**

1. **datos_ejemplo.groupby('id'):** Agrupa el DataFrame datos_ejemplo por la columna 'id'. Esto crea un objeto GroupBy donde los datos se organizan en grupos, cada uno correspondiente a un valor único en la columna 'id'.

2. **['fecha']:** Selecciona la columna 'fecha' del objeto GroupBy. Esto indica que quieres realizar una operación en la columna 'fecha' dentro de cada grupo.

3. **.nunique():** Aplica la función nunique() a la columna 'fecha' de cada grupo. nunique() cuenta el número de valores únicos en una serie. En este caso, cuenta el número de fechas únicas para cada inmueble.

4. **cantidad_dias_por_inmueble = ...:** Asigna el resultado a la variable cantidad_dias_por_inmueble. Esta variable contendrá una serie donde el índice son los IDs de los inmuebles y los valores son la cantidad de días únicos que tienen registros.

```python
datos_ejemplo = datos2.copy()
datos_ejemplo.head()

datos_ejemplo['fecha'] = fecha_ano_mes
datos_ejemplo.head()

cantidad_dias_por_inmueble = datos_ejemplo.groupby('id')['fecha'].nunique()
print(cantidad_dias_por_inmueble)
```
