El m√©todo **apply()** solo funciona para **Series** y el me√©todo **applymap()** para **DataFrame**, una **Serie** es cuando tengo una sola columna o estoy trabajando en una sola columna, pero un **DataFrame** es cuando tengo m√°s de dos columnas

```python
datos['precio'] = datos['precio'].apply(lambda x: x.replace('$','').replace(',','').strip())
```

```python
datos['precio'] = datos['precio'].astype(np.float64)
```

```python
datos.info()
```

para acceder a varias columnas podr√≠a hacerlo de dos formas:

1. creando una lista y despu√©s trabajo con ellas

* colums_moneda = ['cuota_deposito','cuota_limpieza']
* datos[colums_moneda]

2. con el doble par de **[ ]**

* datos[ ['cuota_deposito','cuota_limpieza'] ]

```python
colums_moneda = ['cuota_deposito','cuota_limpieza']
datos[colums_moneda] = datos[colums_moneda].applymap(lambda x: x.replace('$','').replace(',','').strip())
```

```python
datos[colums_moneda] = datos[colums_moneda].astype(np.float64)
```

**Explicaci√≥n al c√≥digo:**

**cantidad_dias_por_inmueble = datos_ejemplo.groupby(' id ')[' fecha '].nunique( )**

1. **datos_ejemplo.groupby('id'):** Agrupa el DataFrame datos_ejemplo por la columna 'id'. Esto crea un objeto GroupBy donde los datos se organizan en grupos, cada uno correspondiente a un valor √∫nico en la columna 'id'.

2. **['fecha']:** Selecciona la columna 'fecha' del objeto GroupBy. Esto indica que quieres realizar una operaci√≥n en la columna 'fecha' dentro de cada grupo.

3. **.nunique():** Aplica la funci√≥n nunique() a la columna 'fecha' de cada grupo. nunique() cuenta el n√∫mero de valores √∫nicos en una serie. En este caso, cuenta el n√∫mero de fechas √∫nicas para cada inmueble.

4. **cantidad_dias_por_inmueble = ...:** Asigna el resultado a la variable cantidad_dias_por_inmueble. Esta variable contendr√° una serie donde el √≠ndice son los IDs de los inmuebles y los valores son la cantidad de d√≠as √∫nicos que tienen registros.

```python
datos_ejemplo = datos2.copy()
datos_ejemplo.head()

datos_ejemplo['fecha'] = fecha_ano_mes
datos_ejemplo.head()

cantidad_dias_por_inmueble = datos_ejemplo.groupby('id')['fecha'].nunique()
print(cantidad_dias_por_inmueble)
```
