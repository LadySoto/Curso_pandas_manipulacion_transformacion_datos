#**Desafío Administración de Condominios**

```python
import pandas as pd
```

```python
ruta_datos = '/content/drive/MyDrive/Colab Notebooks/Ruta: Aprendiendo a hacer ETL (7)/Curso 4-7 Data Science: Pandas: manupulacion y transformacion de datos/Desafio 2: Administracion de condominios/dados_locacao_imoveis.json'
datos = pd.read_json(ruta_datos)
datos.head()
```

```python
datos = pd.json_normalize(datos['dados_locacao'])
datos.head()
```

##**1. Eliminar datos en listas dentro del DataFrame:**

```python
columnas = list(datos.columns)
columnas
```

```python
datos = datos.explode(columnas[1:])
```

```python
datos.reset_index(inplace=True, drop=True)
```

```python
datos.head()
```

##**2. Verificar tipos de datos:**

```python
datos.info()
```

##**3. Identificar las columnas numéricas y transformarlas a la columnas numéricas de tipo numérico:**

```python
datos.head(2)
```

```python
#La columna numérica es 'valor_aluguel'

datos['valor_aluguel'] = datos['valor_aluguel'].apply(lambda x: x.replace('$','').replace(',','').replace('reais','').strip())
```

```python
import numpy as np

datos['valor_aluguel'] = datos['valor_aluguel'].astype(np.float64)
```

```python
datos.info()
```

##**4. Manipulación de un texto:**

```python
datos['apartamento'] = datos['apartamento'].str.replace('(blocoAP)','')
datos.head()
```

##**5. Datos de tiempo:**

```python
# Transformar para el tipo datetime definiendo el formato de fecha como DD/MM/AAAA ('%d/%m/%Y')
datos['datas_combinadas_pagamento'] = pd.to_datetime(datos['datas_combinadas_pagamento'], format='%d/%m/%Y')
datos['datas_de_pagamento'] = pd.to_datetime(datos['datas_de_pagamento'], format='%d/%m/%Y')

datos
```

```python


# Para contribuir en la solución del contexto es posible calcular la diferencia de días
# entre la fecha combinada y la fecha de pagamento con dt.days
datos['atraso'] = (datos['datas_de_pagamento'] - datos['datas_combinadas_pagamento']).dt.days

datos
```

```python
# Calcular el promedio de tiempo de atraso por apartamentos
media_atraso = datos.groupby(['apartamento'])['atraso'].mean()

media_atraso
```
